```python
#CS1300 Project 5: Visualizing Data
#Francisco J Duran

import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

#Importing Data & Initializing Rows
superheroes = pd.read_csv("superhero_data.csv")
df = pd.DataFrame(data = superheroes)
num_rows = df.shape[0]

#Counting Avengers Dead/Alive/ND
dead_avenger = 0
alive_avenger = 0
nd_avenger = 0
for row in range(0, num_rows): 
  if (df["IS_AVENGER"][row]== "YES" and df["ALIVE"][row]== "Deceased Characters"):#First Question
    dead_avenger +=1
  if (df["IS_AVENGER"][row]== "YES" and df["ALIVE"][row]== "Living Characters"):#First Question
    alive_avenger +=1
  if (df["IS_AVENGER"][row]== "YES" and df["ALIVE"][row]== "ND"):#First Question
    nd_avenger +=1

total_avengers = alive_avenger + dead_avenger + nd_avenger

data = np.array([dead_avenger,alive_avenger, nd_avenger])
label = ["# of Dead Avengers", "# of Avengers Alive", "Unknown"]
colors = ["Red", "Blue", "Grey"]
print("Alive: ",alive_avenger)
print("Dead: ",dead_avenger)
print("ND: ",nd_avenger)
print("Total :",total_avengers)
def absolute_value(val):
    a  = np.round(val/100.*data.sum(), 0)
    return a
plt.title("Avengers Alive Based on CSV File", fontsize = 15)
patches = plt.pie(data, labels = label, autopct = absolute_value, colors = colors)
plt.legend(label, loc="lower right")
plt.show
```

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

#Importing Data & Initializing Rows
superheroes = pd.read_csv("superhero_data.csv")
df = pd.DataFrame(data = superheroes)
num_rows = df.shape[0]

#Setting Integers & Arrays
a=b=c=d=i=j=k=l=m=n=o= 0
list = []
listt = []
listtt = []
x = []
y = []
w = []
z = []

#Loop to go through each row
for row in range(0, num_rows): 
  if (df["ID"][row]== "Secret Identity" and df["ALIVE"][row]== "Deceased Characters"):#First Question
    list.append(df["name"][row])
    i+=1
  if (df["FIRST_MONTH"][row]== "August" and df["ALIGN"][row]== "Good Characters"): #Second Question
      listt.append(df["name"][row])
      j+=1
  if (df["FIRST_MONTH"][row]== "August" and df["ALIGN"][row]== "Bad Characters"): #Third Question
      listtt.append(df["name"][row])
      k+=1
  if(df["FIRST_YEAR"][row] != "ND"): #ND not reading so not included
    if(df["UNIVERSE"][row]== "MARVEL" and int(df["FIRST_YEAR"][row]) <= 1955): #Marvel 1935-1995
      l+=1
    if(df["UNIVERSE"][row]== "MARVEL" and int(df["FIRST_YEAR"][row]) <= 1975 and int(df["FIRST_YEAR"][row]) > 1955):#Marvel 1955-1975
      m += 1
    if(df["UNIVERSE"][row]== "MARVEL" and int(df["FIRST_YEAR"][row]) <= 1995 and int(df["FIRST_YEAR"][row]) > 1975):#Marvel 1975-1995
      n += 1
    if(df["UNIVERSE"][row]== "MARVEL" and int(df["FIRST_YEAR"][row]) <= 2015 and int(df["FIRST_YEAR"][row]) > 1995):#Marvel 1995-2015
      o += 1
    if(df["UNIVERSE"][row]== "DC" and int(df["FIRST_YEAR"][row]) <= 1955): #DC 1935-1955
      a+=1
    if(df["UNIVERSE"][row]== "DC" and int(df["FIRST_YEAR"][row]) <= 1975 and int(df["FIRST_YEAR"][row]) > 1955): #DC 1955-1975
      b += 1
    if(df["UNIVERSE"][row]== "DC" and int(df["FIRST_YEAR"][row]) <= 1995 and int(df["FIRST_YEAR"][row]) > 1975): #DC 1975-1995
      c += 1
    if(df["UNIVERSE"][row]== "DC" and int(df["FIRST_YEAR"][row]) <= 2015 and int(df["FIRST_YEAR"][row]) > 1995):#DC 1995-2015
      d += 1

#Showing Results
print("Number of characters that died with a secret identity: ", i)
print("Number of characters created in August that are good: ", j)
print("Number of characters created in August that are bad: ", k)
print("\nCharacters that died having secret identity: ", list)
print("Good Characters made in August: ", listt)
print("Bad Characters made in August: ",listtt)

print("\n\n")



######## Plot 1 ############
y.append(l)
y.append(m)
y.append(n)
y.append(o)

x.append("1935-1955")
x.append("1955-1975")
x.append("1975-1995")
x.append("1995-2015")


plt.bar(x, y, color = 'r', width = 0.72, label = "# of Characters Created")
plt.xlabel('Year Created')
plt.ylabel('# of Marvel Characters Created')
plt.title('Number of Marvel Characters Created Every 20 Years')
plt.legend()
plt.show()
#Data
print("\nMarvel Characters created on and before 1955: ", l)
print("Marvel Characters created after 1955 and on/before 1975: ", m)
print("Marvel Characters created after 1975 and on/before 1995: ", n)
print("Marvel Characters created after 1995 and on/before 2015: ", o, "\n")


######## Plot 2 ############
z.append(a)
z.append(b)
z.append(c)
z.append(d)

w.append("1935-1955")
w.append("1955-1975")
w.append("1975-1995")
w.append("1995-2015")
plt.bar(w, z, color = 'b', width = 0.72, label = "# of Characters Created")
plt.xlabel('Year Created')
plt.ylabel('# of DC Characters Created')
plt.title('Number of DC Characters Created Every 20 Years')
plt.legend()
plt.show()
#Data
print("\nDC Characters created on and before 1955: ", a)
print("DC Characters created after 1955 and on/before 1975: ", b)
print("DC Characters created after 1975 and on/before 1995: ", c)
print("DC Characters created after 1995 and on/before 2015: ", d)
```
